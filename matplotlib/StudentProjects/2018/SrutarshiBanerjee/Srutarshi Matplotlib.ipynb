```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('Petersburg Station Weather Data.csv')
```

```python
data
```

```python
f, ax = plt.subplots(1,figsize=(10,7))
ax.scatter(data['TEMP-LOW'], data['PRECIPITATION'], c = 'blue', label='TEMP-LOW', zorder=1)
ax.scatter(data['TEMP-HIGH'], data['PRECIPITATION'], c = 'red', label='TEMP-HIGH', zorder=2)
ax.set_xlabel('TEMPERATURE', fontsize=20)
ax.set_ylabel('PRECIPITATION',fontsize=20)
ax.legend(loc = 'right', fontsize = 14)
ax.set_title('PRECIPITATION vs TEMPERATURE')

ax.set_xlim(0, 100)
ax.set_ylim(0, 90)
```

```python
f1, ax1 = plt.subplots(1,figsize=(10,7))
ax1.plot(data['TEMP-LOW'],label='TEMP-LOW')
ax1.plot(data['TEMP-HIGH'],label='TEMP-HIGH')
ax1.set_xlabel('DAYS', fontsize=20)
ax1.set_ylabel('TEMPERATURE',fontsize=20)
ax1.legend(loc = 'upper left', fontsize = 14)
ax1.set_title('TEMPERATURE vs DAYS')
ax1.set_xlim(0, 365)
ax1.set_ylim(0, 82)

ax2 = ax1.twinx()
ax2.plot(data['TEMP-HIGH'] - data['TEMP-LOW'], label = 'TEMP Diff', color='g')
ax2.set_ylabel('TEMPERATURE Diff', fontsize=20)
ax2.legend(loc= 'upper right', fontsize=14)
```

```python
f2, ax2 = plt.subplots(1,figsize=(10,7))
ax2.scatter(data['TEMP-LOW'],data['TEMP-HIGH'])
ax2.set_xlabel('TEMP-LOW', fontsize=20)
ax2.set_ylabel('TEMP-HIGH',fontsize=20)
ax2.set_title('TEMP-HIGH vs TEMP-LOW')
```

```python
f3, ax3 = plt.subplots(1,figsize=(10,7))
ax3.plot(data['DLY-TMIN-NORMAL'],label='DLY-TMIN-NORMAL')
ax3.plot(data['DLY-TMAX-NORMAL'],label='DLY-TMAX-NORMAL')
ax3.set_xlabel('DAYS',fontsize=20)
ax3.set_ylabel('CUMULATIVE TEMP',fontsize=20)
ax3.set_title('CUMULATIVE TEMP VS DAYS')
ax3.set_xlim(0, 365)
ax3.set_ylim(0,810)
ax3.legend(loc = 'upper left', fontsize = 14)

ax3a = ax3.twinx()
ax3a.plot(data['DLY-TMAX-NORMAL'] - data['DLY-TMIN-NORMAL'],label='Difference', color = 'g')
ax3a.set_ylabel('TEMPERATURE Diff', fontsize=20)
ax3a.legend(loc= 'upper right', fontsize=14)
```

```python
f4, ax4 = plt.subplots(1,figsize=(10,7))
ax4.plot(data['PRECIPITATION'])
ax4.set_xlabel('DAYS',fontsize=20)
ax4.set_ylabel('PRECIPITATION',fontsize=20)
ax4.set_title('PRECIPITATION VS DAYS')

ax4.set_xlim(0, 365)
ax4.set_ylim(0,365)
```
