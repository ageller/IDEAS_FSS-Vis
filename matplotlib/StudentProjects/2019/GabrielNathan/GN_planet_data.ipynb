```python
#define the subplots and figure size
f, (ax) = plt.subplots(1, figsize=(10,7))

#plot the data with better colors
ax.scatter(data['fst_mass'], data['fst_lum'], c = 'lightblue', edgecolors='darkblue', zorder = 3,label='Observation')
sorteddata=data.sort_values(by='fst_mass')

x=sorteddata['fst_mass']
y=np.log10(sorteddata['fst_mass']**3.5)

ax.set_xlim(-1, 8)
ax.set_ylim(-4, 4)

ax.grid(color='gray', linestyle=':', linewidth=0.5, zorder = 1)
ax.plot([0,0],[-5,5], color='black', zorder = 2)
ax.plot([-2,8],[0,0], color='black', zorder = 2)
#Add analytic prediction
ax.plot(x, y, c = 'red', linestyle='--',linewidth=3,zorder=3,label='Analytic Prediction')

ax.legend()

textstr = '\n'.join((
    r'$\frac{L}{ L_\odot} = {(\frac{M}{ M_\odot})}^{3.5}$',))
props = dict(boxstyle='round', facecolor='red', alpha=0.5)
ax.text(0.7, 0.18, textstr, transform=ax.transAxes, fontsize=25,
        verticalalignment='top', bbox=props)

ax.set_title('Stellar Luminosity versus Mass', fontsize=24)
ax.set_xlabel('Stellar Mass [M$\odot$]', fontsize=20)
ax.set_ylabel('Stellar Luminosity [log10(L$\odot$)]',fontsize=20)
```
