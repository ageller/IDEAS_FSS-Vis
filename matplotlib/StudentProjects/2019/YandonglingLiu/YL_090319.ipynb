```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
c_all = pd.read_csv('../../../../datasets/congress_data.csv', 
                    usecols=['congress_number','x_dimension','alt_dimension','num_votes'])
c114 = c_all[(c_all['congress_number']==114)]
c114
```

```python
#define the subplots and figure size
f, ax = plt.subplots(figsize=(8,7))

#plot the data with better colors
plot = ax.scatter(c114['x_dimension'], c114['alt_dimension'], c=c114['num_votes'], facecolors='None', cmap='viridis', linewidth=1.5)
plot.set_facecolor('None')
ax.set_xlabel('x', fontsize = 20)
ax.set_ylabel('alt',fontsize = 20)
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_title('# Votes in Congress 114',fontsize=24)
t = np.arange(9)/4. - 1
ax.set_yticks(t)
ax.set_xticks(t)
ax.tick_params(labelsize=14)
#ax.grid(color='gray', linestyle='-', linewidth=1, zorder = 1)
ax.plot([0,0],[-2,2], color='black', zorder = 2)
ax.plot([-2,2],[0,0], color='black', zorder = 2)
cbar = plt.colorbar(plot)
cbar.ax.tick_params(labelsize=14)
cbar.ax.set_ylabel('Number of Votes', fontsize=20)
plt.show()

#plt.savefig('scatter.pdf',format='pdf')
```
