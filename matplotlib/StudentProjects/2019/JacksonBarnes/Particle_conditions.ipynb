```python
import numpy as np
import matplotlib; import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import matplotlib.ticker as ticker
import math
```

```python
df = pd.read_csv('ss.00010000.r.bt', delim_whitespace=True, 
    names=('orig ID', 'ID', 'mass', 'radius', 'x', 'y', 'z', 'vx', 'vy', 'vz', 'sx', 'sy', 'sz', 'color'))


#for radial distance from center
df['r'] = np.power((df['x']**2)+(df['y']**2)+(df['z']**2),0.5)
df.hist(column='r', bins=8, edgecolor='black', linewidth=1.0, color = 'c', zorder = 10)
#print(df['r'].min(), bins=8)
#print(df['r'].max())
plt.title('radial distance')
plt.xticks(np.arange(0, 0.00012, 0.00002))
plt.minorticks_on()
plt.tick_params(axis='x', which='minor', length=218, width=0.1, direction='in')
plt.xlabel('r (AU)')
plt.ylabel('number of particles')
#plt.savefig('radial.png')

#for azimuthal angle
df['theta'] = np.arctan(df['y']/df['x'])
df.hist(column="theta", bins=8, edgecolor='black', linewidth=1.0, color = 'Red', zorder = 10)
plt.title('azimuthal angle component')
plt.xticks([-np.pi/2, -3*np.pi/8, -np.pi/4, -np.pi/8, 0, np.pi/8, np.pi/4, 3*np.pi/8, np.pi/2],
          [r'$-\pi/2$', r'$-3\pi/8$', r'$-\pi/4$', r'$-\pi/8$', r'$0$', r'$+\pi/8$', r'$+\pi/4$', r'$+3\pi/8$', r'$+\pi/2$'])
plt.yticks([0, 100, 200, 300, 400, 500, 600, 700, 800])
plt.xlabel('theta (radians)')
plt.ylabel('number of particles')
#plt.savefig('theta.png')


#for polar angle
df['phi'] = np.arccos(df['z']/df['r'])
df.hist(column="phi", bins=8, edgecolor='black', linewidth=1.0, color= 'g', zorder = 10)


plt.title('polar angle component')
plt.xticks([0, np.pi/8, np.pi/4, 3*np.pi/8, np.pi/2, 5*np.pi/8, 3*np.pi/4, 7*np.pi/8, np.pi],
          [r'$0$', r'$\pi/8$', r'$\pi/4$', r'$3\pi/8$', r'$\pi/2$', r'$5\pi/8$', r'$3\pi/4$', r'$7\pi/8$', r'$\pi$'])
plt.yticks([0, 100, 200, 300, 400, 500, 600, 700, 800])
plt.xlabel('phi (radians)')
plt.ylabel('number of particles')
plt.savefig('phi.png')
```
