## Scatter Plot

```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
exo = pd.read_csv('../../../../datasets/compositepars_2019.08.21_11.02.05.csv')
print(exo.fpl_bmasse)
print(exo.fpl_rade)
```

```python
fig, ax2 = plt.subplots(figsize=(10,10))


ax2.plot(exo['fst_mass'],exo['fst_rad'],marker='.',markersize=2,ls='None',color='blue')
ax2.set_xlabel('Mass [Earth Mass]',fontsize=20)
ax2.set_ylabel('Radius [Earth Radii]',fontsize=20)
ax2.axhline(y=1,linestyle='--',color='black')
ax2.axvline(x=1,linestyle='--',color='r',linewidth='1')

plt.tight_layout()
plt.show()
```

### Let's try to involve the orbital distances as well

```python
exo.fpl_smax;
```

```python
import pandas as pd
df = pd.DataFrame(exo['fpl_smax'])
dis_small = df[df['fpl_smax']<0.1]
dis_medium = df[(1>df['fpl_smax'])&(df['fpl_smax']>0.1)]
dis_large =df[df['fpl_smax']>1]
```

```python
df = pd.DataFrame({'Distance' : exo['fpl_smax'], 'Radius' : exo['fst_rad'],'Mass' : exo['fst_mass']})
df
```

#### Suppose we want to get a feeling of the distance distribution with respect to the Radius vs Mass relation while trying to find the most 'Earth like' candidate (distance of 1 AU)
#### We can split this data set into 3 categories:
#### Orbital distance smaller than 0.8 AU => small
#### Orbital distance between 0.8 AU and 1.2 AU (good candidates) => medium
#### Orbital distance between 1.2 AU and 5 AU => large

```python
dis_small = df[df['Distance']<0.8]
dis_medium = df[(1.2>df['Distance'])&(df['Distance']>0.8)]
dis_large =df[(1.2<df['Distance'])&(df['Distance']<5)];
```

```python
fig, ax2 = plt.subplots(figsize=(10,10))


ax2.scatter(np.log(dis_small['Mass']),np.log(dis_small['Radius']),s=100,facecolors='none', edgecolors='r',label='small')
ax2.scatter(np.log(dis_medium['Mass']),np.log(dis_medium['Radius']),s=150,facecolors='none', edgecolors='black',label='medium')
ax2.scatter(np.log(dis_large['Mass']),np.log(dis_large['Radius']),s=200,facecolors='none', edgecolors='b',label='large')
ax2.set_xlabel('Mass [Earth Mass]',fontsize=20)
ax2.set_ylabel('Radius [Earth Radii]',fontsize=20)
ax2.axhline(y=0,linestyle='--',color='brown',linewidth='1')
ax2.axvline(x=0,linestyle='--',color='brown',linewidth='1')
ax2.set_xlim(-2, 2)
ax2.set_ylim(-2, 2)

plt.legend(fontsize=20)
plt.tight_layout()
plt.show()
```

#### It's a bit hard to see what's going on with all this data overlap, so we might as well do 3 different plots

```python
fig, (ax1, ax2,ax3) = plt.subplots(3, 1,figsize=(9,12))

ax1.scatter(np.log(dis_small['Mass']),np.log(dis_small['Radius']),s=100,facecolors='none', edgecolors='r',label='small')
ax1.set_xlabel('Mass [Earth Mass]',fontsize=20)
ax1.set_ylabel('Radius [Earth Radii]',fontsize=20)
ax1.axhline(y=0,linestyle='--',color='brown',linewidth='1')
ax1.axvline(x=0,linestyle='--',color='brown',linewidth='1')
ax1.set_xlim(-2, 2)
ax1.set_ylim(-2, 2)
ax1.set_title('Small Orbital Distances (less than 0.8 AU)',fontsize=20)

ax2.scatter(np.log(dis_medium['Mass']),np.log(dis_medium['Radius']),s=150,facecolors='none', edgecolors='black',label='medium')
ax2.set_xlabel('Mass [Earth Mass]',fontsize=20)
ax2.set_ylabel('Radius [Earth Radii]',fontsize=20)
ax2.axhline(y=0,linestyle='--',color='brown',linewidth='1')
ax2.axvline(x=0,linestyle='--',color='brown',linewidth='1')
ax2.set_xlim(-2, 2)
ax2.set_ylim(-2, 2)
ax2.set_title('Optimal Orbital Distances (between 0.8 AU and 1.2 AU)',fontsize=20)

ax3.scatter(np.log(dis_large['Mass']),np.log(dis_large['Radius']),s=200,facecolors='none', edgecolors='b',label='large')
ax3.set_xlabel('Mass [Earth Mass]',fontsize=20)
ax3.set_ylabel('Radius [Earth Radii]',fontsize=20)
ax3.axhline(y=0,linestyle='--',color='brown',linewidth='1')
ax3.axvline(x=0,linestyle='--',color='brown',linewidth='1')
ax3.set_xlim(-2, 2)
ax3.set_ylim(-2, 2)
ax3.set_title('Large Orbital Distances (between 1.2 AU and 5 AU)',fontsize=20)

plt.tight_layout()
plt.show()
```

### Of course, we can't really say that the middle one will give the best candidates since it depends on the star the planet is orbiting, but at least we get a better idea of what the planets are like, their distance distribution, and their radius vs mass relation
