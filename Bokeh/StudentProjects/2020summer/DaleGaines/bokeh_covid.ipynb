```python
import numpy as np
from bokeh.plotting import *
from bokeh.layouts import row, column
from bokeh.models import ColumnDataSource, CustomJS, Slider, HoverTool
import pandas as pd
output_notebook()
```

```python
# https://covidtracking.com/data/state/illinois
hist = pd.read_csv('illinois-history.csv', parse_dates=['date']).fillna(0)
hist = hist.to_dict(orient='list')
hist
```

```python
source = ColumnDataSource(data=hist)

hover = HoverTool(tooltips=[('Number',   '$y'), ('date', '@date{%F}')],
                  formatters={'date': 'datetime', 'Number': 'printf'})
TOOLS = [hover]

# create a new plot and add a renderer
p1 = figure(tools=TOOLS, width=600, height=350, title='Covid Deaths in Illinois', x_axis_label='Date', x_axis_type='datetime')
p1.line('date', 'death', source=source, color='red', line_width=3)

p2 = figure(tools=TOOLS, width=600, height=350, title='Covid Cases in Illinois', x_axis_label='Date', x_axis_type='datetime')
p2.line('date', 'positive', source=source, color='blue', line_width=3)


# show the results
show(column([p1,p2]))
```
