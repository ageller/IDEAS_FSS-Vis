```python
import pandas as pd
import numpy as np
from astropy import units
from astropy.coordinates import SkyCoord, Distance
from astropy.cosmology import WMAP9
import datetime

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('default')
```

```python
df = pd.read_csv('Rubin_data.csv')

df = df.loc[(df['Days_Since_Start'].notnull()) & \
              (df['RA'].notnull()) & \
              (df['Dec'].notnull()) & \
              (df['skybrightness'].notnull()) & \
              (df['moonphase'].notnull()) & \
              (df['filter'].notnull())] 
df
```

```python
x = []
y = []
z = []


distance = 100#arbitrary distance 
c1 = SkyCoord(df['RA'], df['Dec'], unit=(units.hourangle, units.deg), distance=distance*units.Mpc).galactic.cartesian

x = (c1.x.to(units.Mpc).value)
y = (c1.y.to(units.Mpc).value)
z = (c1.z.to(units.Mpc).value)

```

```python
obsv = {
        'Days_Since_Start': df['Days_Since_Start'],
        'x': x,
        'y': y,
        'z': z,
        'skybrightness': df['skybrightness'],
        'moonphase': df['moonphase'],
        'cloud': df['cloud'],
        'airmass':df['airmass'],
        'filter':df['filter']
        }

df = pd.DataFrame(obsv, columns= [ 'Days_Since_Start','x','y','z','skybrightness','moonphase','cloud','airmass','filter'])


df 
```
