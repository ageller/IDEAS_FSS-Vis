```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib
import imageio
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
filename = 'TUDCrossing.mp4'
reader = imageio.get_reader(filename, 'ffmpeg')

for i, im in enumerate(reader):
    if i == 0:
        im1 = resize(im, (480,480,3))
        im2 = resize(im1, (512, 512, 3))
        print(im2.shape)
        plt.imshow(im2)
        imageio.imwrite('TUDCrossing_Frame1.jpg',im2)
```
