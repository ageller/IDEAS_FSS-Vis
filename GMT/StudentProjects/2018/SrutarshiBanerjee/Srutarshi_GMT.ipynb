```python
import gmt
fig = gmt.Figure()
fig.coast(region=[95, 160, -12, 25], projection='M6i', land='blue', 
          frame=True)
fig.show()
```

```python
quakes = gmt.datasets.load_usgs_quakes()
fig = gmt.Figure()
grid = gmt.datasets.load_earth_relief(resolution="10m")
fig.basemap(region=[95, 160, -12, 25], projection="M10i", frame=True)
fig.grdimage(grid, cmap="geo")
fig.plot(x=quakes.longitude, y=quakes.latitude, 
         sizes=0.005*2**quakes.mag, color=quakes.depth/quakes.depth.max(),
         style="ci", pen="yellow", cmap="yellow")

fig.savefig('asia-pacific_earthquakes.png')
fig.show()
```
